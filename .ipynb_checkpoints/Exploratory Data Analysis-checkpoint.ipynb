# Exploratory Data Analysis

Before diving into the Exploratory Data Analysis, I like to make some assumptions. I need to find out what variables are important when we predict the price of an Airbnb in Paris and what are there effects on the Price. These are my assumptions:
- 

# Exploring the feature year_last_review

```python
airbnb.year_last_review.describe()
```

```python
# That's weird some apartments have reviews from 1970. Let us explore these apartments.
airbnb[airbnb.year_last_review==1970]
```

```python
# Airbnb was created in 2008 therefore, this is not possible. Let's drop thesde apartments.
airbnb1=airbnb[airbnb.year_last_review!=1970]
```

```python
airbnb1.year_last_review.describe()
```

Due to the increase in the price of apartments in Paris between 2010 to 2020, apartments that have n ot any reviews after 2018 are not relevant.

```python
airbnb2=airbnb[airbnb.year_last_review>=2018]
```

```python
airbnb2.shape
```

# Exploring the feature Price

```python
airbnb2.price.describe()
```

```python
# It seems that some apartments are very expensive. Let us explore them.
airbnb2[airbnb2.price>4000]
```

After looking on the airbnb website in pAris we did not find any apartments with a price higher than 4000 euros. Therefore we will drop the apartments with a price higher than 4000 euros. We consider them as outliers.


```python
airbnb3=airbnb2[airbnb2.price<4000]
```

Same thing, after looking on the airbnb website in Paris we did not find any apartments with a price lower than 45 euros. Therefore we will drop the apartments with a price lower than 45 euros. We consider them as outliers.

```python
airbnb4=airbnb3[airbnb3.price>45]
airbnb4.shape
```

```python
airbnb4.describe()
```

# Let's dive into the feature minimum_nights

```python
#After looking on airbnb website, we have seen that it is now possible to rent an apartment for more than 28 days.
airbnb4[airbnb4.minimum_nights>28]
```

```python
# Why because not normal usage of airbnb
airbnb5=airbnb4[airbnb4.minimum_nights<=28]
airbnb5.shape
```

```python
airbnb5.describe()
```
