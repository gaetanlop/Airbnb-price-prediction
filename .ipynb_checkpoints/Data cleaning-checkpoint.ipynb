# Data Cleaning

## Transform the dataframe reviews so that it indicates the number of reviews per listing_id.

```python
reviews.head(10)
```

```python
pd.DataFrame(reviews.listing_id.value_counts())
```

```python
number_of_reviews=pd.DataFrame(reviews.listing_id.value_counts())
```

```python
number_of_reviews=number_of_reviews.reset_index().rename(columns={'index': 'id', 'listing_id': 'reviews_number'})
```

```python
number_of_reviews
```

```python
pd.merge()
```

## Merge the datasets

```python
airbnb=listings.merge(number_of_reviews,on='id', how='left')
```

```python
airbnb.info()
```

```python
airbnb.isnull().sum()
```

There are 15261 missing values in 'last_review','reviews_per_month','reviews_number'. It is obvious that these values are missing because there are no reviews on this airbnb. therefore we will replace those values by 0.

```python
airbnb['reviews_number'].fillna(0,inplace=True)
airbnb['last_review'].fillna(0,inplace=True)
airbnb['reviews_per_month'].fillna(0,inplace=True)
```

```python
airbnb.isnull().sum()
```

The column neighbourhood_group is full of missing values. Therefore we will drop it.

```python
#Dropping the column neighbourhood_group
airbnb1=airbnb.drop('neighbourhood_group',axis=1)
```

```python
airbnb1.isnull().sum()
```

For our analmysis we do not need any informations about the host name. This column will not have any impact on the price of a airbnb. Thus, we will drop them.

```python
airbnb2=airbnb1.drop('host_name',axis=1)
```

```python
# Dropping missing values of the column 'name'. This column corresponds to the name of the airbnb given by the host.
airbnb2.dropna(axis=0,inplace=True)
```

```python
airbnb2.isnull().sum()
```

```python
airbnb2.shape
```

```python
airbnb2.head()
```

```python
airbnb2.drop('id',axis=1,inplace=True)
```

```python
airbnb2.head()
```

```python
df=pd.DataFrame(airbnb2.host_id.value_counts())
def count(x):
    return df.loc[x].host_id   
    
```

```python
airbnb2['number_of_apartments']=airbnb2['host_id'].apply(lambda x: count(x))
```

```python
airbnb2.head()
```

```python
airbnb2['last_review']=pd.to_datetime(airbnb2['last_review'])
```

```python
airbnb2['year_last_review'] = pd.DatetimeIndex(airbnb2['last_review']).year
```

```python
airbnb2.head()
```

```python
airbnb2.drop('last_review',axis=1,inplace=True)
```

```python
airbnb2.to_csv('airbnb_cleaned.csv',index=False)
```
